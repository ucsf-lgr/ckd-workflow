```python
# Creating sample sequences
ps1_max_scores = []
ps2_max_scores = []
cs1_rc = Seq(CS1).reverse_complement()

for (ps1, ps2) in zip(PS1s, PS2s):
    # Finding similarities
    ps1 = Seq(ps1)
    ps2 = Seq(ps2)

    ps1_alignments = pairwise2.align.globalms(cs1_rc, ps1, 2, -1, -10, -10, penalize_end_gaps=False)
    ps2_alignments = pairwise2.align.globalms(cs1_rc, ps2, 2, -1, -10, -10, penalize_end_gaps=False)
    
    # Showing results
    ps1_scores    = [x.score for x in ps1_alignments]
    ps2_scores    = [x.score for x in ps2_alignments]
    ps1_max_scores += [max(ps1_scores)]
    ps2_max_scores += [max(ps2_scores)]

    #print(ps1_alignments[1])
    for a in ps1_alignments:
        print(format_alignment(*a))
    for a in ps2_alignments:
        print(format_alignment(*a))
    print("***************************************")
    
```
